## Importing the Libraries

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns
```

## Obtaining Probability Curves

```python
# Parameters for the normal distribution
mu = 1000      # Mean
sigma = 20     # Standard deviation
n_samples = 10000  # Number of samples

# Variables for the threshold values
limite_a = 1024  # Upper threshold
limite_b = 961   # Lower threshold

# Generate n_samples from a normal distribution N(mu, sigma)
samples = np.random.normal(mu, sigma, n_samples)

# Calculate the empirical (computed) probabilities
p_computational_a = np.mean(samples < limite_a)
# Probability that X is less than limite_a
p_computational_b = np.mean((samples < limite_a) & (samples > limite_b)) / np.mean(samples > limite_b)
# Conditional probability P(X < limite_a | X > limite_b)

# Calculate the theoretical probabilities
p_theoretical_a = norm.cdf(limite_a, mu, sigma)
# Theoretical probability that X is less than limite_a
p_theoretical_b = (norm.cdf(limite_a, mu, sigma) - norm.cdf(limite_b, mu, sigma)) / (1 - norm.cdf(limite_b, mu, sigma))
# Theoretical conditional probability P(X < limite_a | X > limite_b)

# Plot histogram and probability P(X < limite_a)
plt.figure(figsize=(14, 6))

# Plot for P(X < limite_a)
plt.subplot(1, 2, 1)
plt.hist(samples, bins=50, density=True, alpha=0.6, color="skyblue", edgecolor="black")
# Plot the histogram of samples with density normalized
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)  # X-axis range for plotting
plt.plot(x, norm.pdf(x, mu, sigma), 'r', label="Theoretical Distribution")
# Overlay the theoretical normal distribution

# Fill the area under the curve for P(X < limite_a)
x_fill = np.linspace(mu - 4*sigma, limite_a, 1000)  # Range up to limite_a
plt.fill_between(x_fill, norm.pdf(x_fill, mu, sigma), color="lightcoral", alpha=0.6, label=f"P(X < {limite_a})")
# Fill area for probability calculation

plt.title(f"Distribution of X and P(X < {limite_a})\nComputed Prob: {p_computational_a:.5f}, Theoretical Prob: {p_theoretical_a:.5f}")
plt.xlabel("Values of X")
plt.ylabel("Probability Density")
plt.legend()

# Plot for P(X < limite_a | X > limite_b)
plt.subplot(1, 2, 2)
plt.hist(samples, bins=50, density=True, alpha=0.6, color="skyblue", edgecolor="black")
# Histogram of samples
plt.plot(x, norm.pdf(x, mu, sigma), 'r', label="Theoretical Distribution")
# Overlay the theoretical normal distribution

# Fill the area under the curve for conditional probability P(X < limite_a | X > limite_b)
x_fill_b = np.linspace(limite_b, limite_a, 1000)  # Range from limite_b to limite_a
plt.fill_between(x_fill_b, norm.pdf(x_fill_b, mu, sigma), color="lightgreen", alpha=0.6, label=f"P(X < {limite_a} | X > {limite_b})")
# Fill area for conditional probability

plt.title(f"Distribution of X and P(X < {limite_a} | X > {limite_b})\nComputed Prob: {p_computational_b:.5f}, Theoretical Prob: {p_theoretical_b:.5f}")
plt.xlabel("Values of X")
plt.ylabel("Probability Density")
plt.legend()

plt.tight_layout()
plt.show()
```

The above images show two probability distributions for an unknown variable X, with each graph displaying the probability density and a comparison between theoretical and computed probabilities under different conditions.

1. Left (Distribution of X and Probability of X less than 1024):
- The first figure on the left shows the distribution of X along with the
conditional probability that X is less than 1024.
- The computed probability of X being less than 1024 is 0.88540, while the expected theoretical probability is 0.88493.
- The difference between the computed and theoretical values is very small, only 0.00047, indicating a good agreement between the simulation and theory.


2. Right (Distribution of X and Probability of X less than 1024 given X greater than 961):
- The second figure on the right represents the distribution of X, considering only the condition that X is greater than 961, and the conditional probability of X being less than 1024 given X greater than 961.
- The computed probability is 0.88234, while the theoretical probability is 0.88191.
- The difference here is also very small, around 0.00043, showing a strong agreement between the theoretical and computed values, even under the additional condition of X being greater than 961.


3. Comparison Between Both Graphs:
- Both distributions show a high degree of precision between the computed and theoretical values, indicating that the probabilistic model used for the simulations is robust.
- The small difference between the computed and theoretical values in both graphs demonstrates that the calculation and simulation methods are well-tuned to model the variable X.
- It is observed that the second figure, which considers the condition of X being greater than 961, slightly shifts the curve to the right, restricting the conditional probability area compared to the first. This adjustment represents an informative update, as it conditions the probability on the upper range of X.
- This comparison can be useful in practical scenarios, where additional conditions or constraints on random variables may slightly alter observed probabilities without drastically affecting the overall distribution behavior.

These analyses confirm that the simulations provide a good theoretical approximation and that additional conditions influence the observed probability in a controlled manner, with minimal impact on the final calculated values. The precision found suggests that the model is reliable for probabilistic inferences about X in both scenarios.

## Analysis of Uniform, Normal and Exponential Distributions With Different Amounts of Samples

```python
# Function to generate and analyze random variables
def analyze_random_variables(n_samples):
    # Generate random variables
    uniform_samples = np.random.uniform(0, 1, n_samples)
    normal_samples = np.random.normal(loc=0, scale=1, size=n_samples)
    exponential_samples = np.random.exponential(scale=1, size=n_samples)

    # Calculate mean and variance
    mean_uniform = np.mean(uniform_samples)
    variance_uniform = np.var(uniform_samples)

    mean_normal = np.mean(normal_samples)
    variance_normal = np.var(normal_samples)

    mean_exponential = np.mean(exponential_samples)
    variance_exponential = np.var(exponential_samples)

    # Theoretical values
    theoretical_mean_uniform = 0.5  # For Uniform(0, 1)
    theoretical_variance_uniform = 1/12  # For Uniform(0, 1)

    theoretical_mean_normal = 0  # For Normal(0, 1)
    theoretical_variance_normal = 1  # For Normal(0, 1)

    theoretical_mean_exponential = 1  # For Exponential(1)
    theoretical_variance_exponential = 1  # For Exponential(1)

    # Print calculated values
    print(f"Samples: {n_samples}")
    print("Uniform Distribution: ")
    print(f"Mean: {mean_uniform:.4f}, Variance: {variance_uniform:.4f} (Expected Mean: {theoretical_mean_uniform:.4f}, Expected Variance: {theoretical_variance_uniform:.4f})")
    print("Normal Distribution: ")
    print(f"Mean: {mean_normal:.4f}, Variance: {variance_normal:.4f} (Expected Mean: {theoretical_mean_normal:.4f}, Expected Variance: {theoretical_variance_normal:.4f})")
    print("Exponential Distribution: ")
    print(f"Mean: {mean_exponential:.4f}, Variance: {variance_exponential:.4f} (Expected Mean: {theoretical_mean_exponential:.4f}, Expected Variance: {theoretical_variance_exponential:.4f})")
    print("\n" + "-"*50 + "\n")

    # Plot histograms and boxplots
    plt.figure(figsize=(18, 12))

    # Uniform
    plt.subplot(3, 2, 1)
    plt.hist(uniform_samples, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
    plt.title('Uniform Distribution Histogram')
    plt.xlabel('Value')
    plt.ylabel('Frequency')

    plt.subplot(3, 2, 2)
    sns.boxplot(y=uniform_samples, color='lightgreen')
    plt.title('Uniform Distribution Boxplot')

    # Normal
    plt.subplot(3, 2, 3)
    plt.hist(normal_samples, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
    plt.title('Normal Distribution Histogram')
    plt.xlabel('Value')
    plt.ylabel('Frequency')

    plt.subplot(3, 2, 4)
    sns.boxplot(y=normal_samples, color='lightgreen')
    plt.title('Normal Distribution Boxplot')

    # Exponential
    plt.subplot(3, 2, 5)
    plt.hist(exponential_samples, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
    plt.title('Exponential Distribution Histogram')
    plt.xlabel('Value')
    plt.ylabel('Frequency')

    plt.subplot(3, 2, 6)
    sns.boxplot(y=exponential_samples, color='lightgreen')
    plt.title('Exponential Distribution Boxplot')

    plt.tight_layout()
    plt.show()

# Analyze for 1,000 samples
analyze_random_variables(1000)

# Analyze for 10,000 samples
analyze_random_variables(10000)

# Analyze for 100,000 samples
analyze_random_variables(100000)
```

To facilitate analysis, we can examine the impacts of increasing the number of samples in four areas: [1] convergence of mean and variance, [2] reduction in variability, [3] histogram smoothing, and [4] reduction of outliers. Thus, we have:

1. Convergence of Sample Mean and Variance to Theoretical Values:
- With more samples, the calculated mean and variance for each distribution (uniform, normal, exponential) approach their respective theoretical values, due to the Law of Large Numbers.
- For example, for a uniform distribution between 0 and 1, the theoretical mean is 0.5, and the theoretical variance is approximately 0.0833. As n_samples increases, the calculated mean and variance converge to these values.

2. Reduction in Sample Statistics Variability:
- With more samples, the variability (or uncertainty) in the calculated mean and variance decreases. This means that the values will fluctuate less and stabilize closer to the theoretical values.
- This phenomenon occurs due to the Central Limit Theorem, which states that, with a sufficiently large sample size, the distribution of the sample mean tends toward a normal distribution centered on the theoretical mean.

3. Smoother Histograms:
- The histogram of the uniform distribution will look more like a straight line.
- The histogram of the normal distribution will approximate the shape of a bell curve.
- The histogram of the exponential distribution will show the exponential decay more accurately.

4. Reduction of Outliers in Boxplots:
- As the number of samples increases, the boxplots for each distribution will better represent the underlying distribution and may appear to have fewer extreme values, as these values become statistically less significant in a larger sample.

In summary, by increasing n_samples, you are enhancing the accuracy and reliability of the empirical estimates of the distributions' means and variances, as well as improving the visual quality of the histograms and boxplots. This is useful in statistical analysis, as larger sample sizes generally provide more accurate insights into the characteristics of the population distribution.
